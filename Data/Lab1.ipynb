```python
# PART F
# cds2aa.py
# Replace all T's with U's
seq_PCCB = seq_PCCB.replace('T', 'U')
# Initialize cds
AAseq = ''
# Loop through sequence
for i in range(0, len_PCCB-3, 3) :
    codon = seq_PCCB[i:i+3]
    AAseq = AAseq + aa[codon]
    
print(AAseq)
```

```python
# Part G 
# lingo.py
AAseq = ''
countAA = dict()
for i in range(0, len_PCCB-3, 3) :
    codon = seq_PCCB[i:i+3]
    AAseq = AAseq + aa[codon]
    if aa[codon] in countAA :
        countAA[aa[codon]] += 1
    else :
        countAA[aa[codon]] = 1
       
valList = list(countAA.values())
keyList = list(countAA.keys())

pos = valList.index(max(valList))

highFreq = keyList[pos]

print('Count of most frequent amino acid ', keyList[pos], ' is: ', valList[pos])

countCodons = dict()
for i in range(0, len_PCCB-3, 3) :
    codon = seq_PCCB[i:i+3]
    if aa[codon] == keyList[pos] :
        if codon in countCodons :
            countCodons[codon] += 1
        else :
            countCodons[codon] = 1
            
for key, value in countCodons.items() :
    print('Count of codon ', key, ' is: ', value)
```
